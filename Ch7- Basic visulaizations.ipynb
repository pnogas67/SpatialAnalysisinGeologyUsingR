###  <span style="color:green;">7.1 Bar plots</span>

```r
#07-01
# Generate sample data
categories = c("A", "B", "C", "D")
values = c(10, 20, 30, 40)

# Create bar plot
barplot(values, names.arg=categories)
```

###  <span style="color:green;">7.2 Histograms</span>

```r
#07-02
# Generate sample data
values = rnorm(100)

# Create histogram
hist(values)
```

###  <span style="color:green;">7.3 Box plots</span>

```r
#07-03
# Generate sample data
values = rnorm(100)

# Create box plot
boxplot(values)
```

###  <span style="color:green;">7.4 Scatter plots</span>

```r
#07-04
# Generate sample data
x = rnorm(100)
y = rnorm(100)

# Create scatter plot
plot(x, y, type='p')
```

###  <span style="color:green;">7.5 Organizing plots</span>

```r
#07-05
# Create a 2x2 layout of plots using par()
par(mfrow = c(2, 2))

# Plot four histograms
hist(rnorm(100), main = "Histogram 1")
hist(rnorm(100), main = "Histogram 2")
hist(rnorm(100), main = "Histogram 3")
hist(rnorm(100), main = "Histogram 4")
```

```r
#07-06
# Create a custom layout of plots using layout()
layout(matrix(c(1, 1, 2, 3), nrow = 2, ncol = 2, byrow = TRUE))

# Plot three histograms
hist(rnorm(100), main = "Histogram 1")
hist(rnorm(100), main = "Histogram 2")
hist(rnorm(100), main = "Histogram 3")
```

###  <span style="color:green;">7.6 Save the plots</span>

```r
# Reset the plot to a 1x1 graph
par(mfrow = c(1, 1))

# Create the device to save the plot to a PNG file
png("my_plot.png")

# create random data
x = rnorm(50)
y = rnorm(50)

# Plot the result
plot(x, y, main = "My Scatterplot", xlab = "X", ylab = "Y")

# Close the graphical device
dev.off()
```
